```python
import pandas as pd
import plotly.express as px
import requests
from io import StringIO

# 1. Catch data
url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"
header = {
  "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
r = requests.get(url, headers=header)

tables = pd.read_html(StringIO(r.text))

#2. Find "World"
df = None
for t in tables:
    if 'Country/Territory' in str(t.columns):
        df = t
        break

if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(-1)

df = df.iloc[:, [0, 1, 2]]
df.columns = ['Country', 'Region', 'GDP']

df['GDP'] = df['GDP'].astype(str).str.replace(r'\[.*\]', '', regex=True)
df['GDP'] = df['GDP'].str.replace(',', '').str.replace('—', '0')
df['GDP'] = pd.to_numeric(df['GDP'], errors='coerce')

df = df.dropna(subset=['GDP'])
df = df[df['Country'] != 'World']

# 3. Picture
fig_gdp = px.bar(df, x="Region", y="GDP", color="Country",
                 title="GDP by Country Stacked within Regions (IMF)",
                 labels={"GDP": "GDP (Millions of USD)"})

fig_gdp.write_html("stacked_bar.html")
fig_gdp.show()
```

```python
import plotly.graph_objects as go

# 1. "ICV", "Brain", "CSF", "Cerebrum", "Cerebellum", "Brainstem", "L-Cerebrum", "R-Cerebrum"
nodes = ["ICV", "Brain", "CSF", "Cerebrum", "Cerebellum", "Brainstem", "L-Cerebrum", "R-Cerebrum"]

# 2.
# ICV (0) -> Brain (1) and CSF (2)
# Brain (1) -> Cerebrum (3), Cerebellum (4), Brainstem (5)
# Cerebrum (3) -> L-Cerebrum (6) 和 R-Cerebrum (7)

source = [0, 0, 1, 1, 1, 3, 3]
target = [1, 2, 3, 4, 5, 6, 7]
values = [1400, 150, 1200, 150, 50, 600, 600]

# 3. Create Figure
fig_sankey = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = nodes,
      color = "blue"
    ),
    link = dict(
      source = source,
      target = target,
      value = values
    ))])

fig_sankey.update_layout(title_text="MRICloud Brain Volume Hierarchy (Sankey Diagram)", font_size=12)

# 4. HTML
fig_sankey.write_html("sankey.html")

# Show figure
fig_sankey.show()
```
